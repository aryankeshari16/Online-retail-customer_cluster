```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
df=pd.read_excel('Online Retail.xlsx')
df.head()
```

```python
#code for removing the letter starts with 'c' it indicates a cancellation. 
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df
```

```python
df['InvoiceDate_Date'] = df['InvoiceDate'].dt.date
df['InvoiceDate_Time'] = df['InvoiceDate'].dt.time
```

```python
#Code to create new feature called TotalSpend which has Total expenditure by every customer
df['TotalSpend'] = df['Quantity'] * df['UnitPrice']
df
```

```python
# Drop rows with missing CustomerID or other critical columns
df_cleaned = df.dropna(subset=['CustomerID', 'Quantity', 'UnitPrice'])

# Convert 'InvoiceDate' to datetime if not already, using .loc to avoid SettingWithCopyWarning
df_cleaned.loc[:, 'InvoiceDate'] = pd.to_datetime(df_cleaned['InvoiceDate'], errors='coerce')

# Drop rows with invalid InvoiceDate
df_cleaned = df_cleaned[df_cleaned['InvoiceDate'].notna()]

# Filter out rows with negative or zero quantities
df_cleaned = df_cleaned[df_cleaned['Quantity'] > 0]

# Create date and time features from InvoiceDate
df_cleaned['InvoiceDate_Date'] = df_cleaned['InvoiceDate'].dt.date
df_cleaned['InvoiceDate_Time'] = df_cleaned['InvoiceDate'].dt.time

# Create a new feature: TotalSpend per transaction
df_cleaned['TotalSpend'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']

# Aggregate customer-level features
customer_features = df_cleaned.groupby('CustomerID').agg({
    'InvoiceNo': 'nunique',          # Frequency
    'TotalSpend': 'sum',             # Monetary
    'Quantity': 'sum'                # For AvgBasketSize
}).rename(columns={
    'InvoiceNo': 'Frequency',
    'TotalSpend': 'Monetary',
    'Quantity': 'TotalQuantity'
})

# Derived customer-level metrics
customer_features['AvgOrderValue'] = customer_features['Monetary'] / customer_features['Frequency']
customer_features['AvgBasketSize'] = customer_features['TotalQuantity'] / customer_features['Frequency']

# Drop the intermediate TotalQuantity column
customer_features.drop(columns='TotalQuantity', inplace=True)

# Merge customer-level features back to cleaned dataset
df_cleaned = df_cleaned.merge(customer_features, on='CustomerID', how='left')

# Display the resulting dataframe
print(df_cleaned.head())
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_features)
```

```python
silhouette_scores = []
range_n_clusters = list(range(2, 11))  # Trying k from 2 to 10

for k in range_n_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_scores.append(score)

# Plotting the Silhouette Scores
plt.figure(figsize=(8, 6))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title("Silhouette Scores for Different Values of k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()
```

```python
# Get optimal k
optimal_k = 3

# Fit KMeans again with optimal k
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_features)

# Assign cluster labels to the same DataFrame used to compute features
customer_features['Cluster'] = cluster_labels
```

```python
df_cleaned = df_cleaned.merge(customer_features['Cluster'], on='CustomerID', how='left')
df_cleaned
```

```python
from sklearn.decomposition import PCA
import seaborn as sns

# Reduce to 2D for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)
customer_features['PCA1'] = pca_result[:, 0]
customer_features['PCA2'] = pca_result[:, 1]

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_features, x='PCA1', y='PCA2', hue='Cluster', palette='tab10', s=60)
plt.title('Customer Segments (PCA Projection)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
